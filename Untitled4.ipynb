```python
X= titanic_df.drop(['survived'], axis=1).values
y=titanic_df['survived'].values
```

```python
X_train, X_test, y_train, y_test= cross_validation.train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
claf_Dec_tree= tree.DecisionTreeClassifier(max_depth=10)
```

```python
claf_Dec_tree
```

```python
claf_Dec_tree.fit(X_train, y_train)
claf_Dec_tree.score(X_test, y_test)
```

```python

shuffle_validator= cross_validation.ShuffleSplit(len(X), n_iter=20, test_size=0.2,random_state=0)
```

```python
Scores_dt= cross_validation.cross_val_score(claf_Dec_tree, X,y, cv=shuffle_validator)
```

```python
print("Accuracy : %0.6f (+/- %0.2f )" % (Scores_dt.mean(), Scores_dt.std()))
```

```python
claf_RF= ske.RandomForestClassifier(n_estimators=50)
```

```python
Scores_RF= cross_validation.cross_val_score(claf_RF, X,y, cv=shuffle_validator)
```

```python
print("Accuracy : %0.6f (+/- %0.2f )" % (Scores_RF.mean(), Scores_RF.std()))
```

```python
claf_GB= ske.GradientBoostingClassifier(n_estimators=50)
```

```python
Scores_GB= cross_validation.cross_val_score(claf_GB, X,y, cv=shuffle_validator)
```
